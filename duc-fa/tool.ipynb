##1 n-gram




###<div dir=rtl>1.1 توضیح
</div>

<div dir=rtl>

N-gram ها دنباله ای پیوسته از کلمات یا نمادها یا نشانه ها در یک سند هستند.

خلاصه، 'n' فقط یک متغیر است که می تواند مقادیر صحیح مثبت از جمله 1،2،3 و غیره داشته باشد.'n اساساً به چندگانه اشاره دارد.

با توجه به همین خطوط، n-gram ها بسته به مقداری که «n» می گیرد، به انواع زیر طبقه بندی می شوند.

n ترم:
</div>


1.  1	Unigram
2.	Bigram
3.	Trigram

n.	n-gram


“this is a blue ball.

SL.No.	---   Type of n-gram	  ---  Generated n-grams
1.	Unigram	: [““this”, "is",  "a", "blue",   "ball"]
2.	Bigram :	[“this is",   ”is a","a blue",   "blue ball"]
3.	Trigram	: ["this is a",   "is a blue",  "a blue ball"]

###1.2 کد

```python
def ngrams(text, n, ignore_split = False):

  if ignore_split:

    """
      ngrams("alireza", 3, True)
      --->['ali', 'lir', 'ire', 'rez', 'eza']
    """
    return [text[index : index + n] for index in range(0, len(text) - n + 1)]


  else:
    """
      ngrams("this is a ball", 2)
      --->[('this', 'is'), ('is', 'a'), ('a', 'ball')]
    """

    text = text.lower()
    words = text.split()
    return [tuple(words[index : index + n]) for index in range(0, len(words) - n + 1)]
```

###1.3 مثال

```python
ngrams("this is a ball", 2)
```

```python
ngrams("alireza", 3, True)
```

##2 counter elements



###2.1 توضیح


<div dir=rtl>

اگر بخواهیم عناصر یک دیکشنری را بشماریم از این تابع ساده زیر استفاده می کنیم. این کار عمدتاً در الگوریتم ها استفاده می شود.


###2.2 کد

```python
def counter_element(temp):

  """

    This function is used to count the elements in n_grams
    temp = [1, 2, 3, 2, 1, 1, 1]
    counter_element(temp)
    ---> {1 : 4, 2 : 2, 3 : 1}

  """

  counter = {}
  for element in temp:
    counter[element] = counter[element] + 1 if element in counter else 1
  return counter
```

###1.2.3 مثال

```python
bigram = ngrams("this is a ball and this is a door", 2)
bigram
```

```python
counter_element(bigram)
```

##1.3 overlaps - dictionary



###1.3.1 توضیح

<div dir=rtl>

در اینجا مشترکات بین 2 دیکشنری را محاسبه می کنیم روش کار به این صورت است که اگر کلیدی در هر دو دیکشنری وجود داشته باشد به عنوان یک عنصر مشترک در نظر گرفته می شود و  مقدار، کمترین مقدار هر یک از دیکشنری ها است. این کار عمدتاً در الگوریتم ها استفاده می شود.

###1.3.2 code

```python
def overlaps_dict(dict1, dict2):

  """
    To find commonalities of 2 dictionaries
    dict1 = {"a" : 3, "b" : 2, "c" : 7}
    dict2 = {"a" : 1, "b" : 5, "f" : 17}
    overlaps_dict(dict1, dict2)
    ---> {"a" : 1, "b" : 2}

  """

  overlaps = dict()
  for key in dict1:
    if key in dict2:
      _val_overlaps = min(dict1[key], dict2[key])
      overlaps.update({key : _val_overlaps})
  return overlaps
```

###1.3.3 example

```python
bigram_1 = ngrams("this is a ball and this is a door", 2)
bigram_1
```

```python
bigram_2 = ngrams("this is a blue ball ", 2)
bigram_2
```

```python
ce_1 = counter_element(bigram_1)
ce_1
```

```python
ce_2 = counter_element(bigram_2)
ce_2
```

```python
overlaps_dict(ce_1, ce_2)
```
